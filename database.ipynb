```python
import sqlite3
```

```python
conn = sqlite3.connect("students.db")
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    age INTEGER,
    studytime INTEGER,
    failures INTEGER,
    absences INTEGER,
    G1 INTEGER,
    G2 INTEGER,
    predicted_G3 REAL
)
""")

conn.commit()
conn.close()

print("Database created successfully!")
```

```python
conn = sqlite3.connect("students.db")
cursor = conn.cursor()

cursor.execute("""
INSERT INTO students (
    age,
    studytime,
    failures,
    absences,
    G1,
    G2,
    predicted_G3
)
VALUES (?, ?, ?, ?, ?, ?, ?)
""", (18, 2, 0, 4, 15, 14, 15.8))

conn.commit()

print("Data inserted!")

conn.close()
```
